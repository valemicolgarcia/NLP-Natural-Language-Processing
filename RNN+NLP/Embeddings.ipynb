```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Load the digits dataset
digits = load_digits()
data = digits.data
targets = digits.target

# Visualize some of the digits
fig, axes = plt.subplots(1, 10, figsize=(10, 3))
for i, ax in enumerate(axes):
    ax.set_axis_off()
    ax.imshow(digits.images[i], cmap='gray')
plt.show()
```

```python
data.shape
```

```python
import pandas as pd
pd.DataFrame(data)
```

# Mapeamos a 2D

```python
from sklearn.manifold import TSNE

# Perform t-SNE embedding
tsne = TSNE(n_components=2, random_state=42)
embedding = tsne.fit_transform(data)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x=embedding[:, 0],y= embedding[:, 1], hue=targets,palette="deep")
```

# Word embeddings

```python
import gensim.downloader as api

# Load the pre-trained Word2Vec model
model = api.load("word2vec-google-news-300")
#test glove-wiki-gigaword-50
```

```python
model['man']
```

```python
model['man'].shape
```

```python
man = model['man']
woman = model['woman']
```

```python
#woman + genre = man
```

```python
genre = man - woman
```

```python
aunt = model['aunt']
```

```python
x = aunt + genre
```

```python
similar_words = model.most_similar(x, topn=10)
```

```python
for word, similarity in similar_words[1:]:
    print(f"{word}: {similarity:.4f}")
```

```python
hija = model['daughter']
```

```python
similar_words = model.most_similar(hija+genre, topn=10)
for word, similarity in similar_words[:]:
    print(f"{word}: {similarity:.4f}")
```
