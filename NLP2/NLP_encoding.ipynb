```python
import pandas as pd

# Synthetic dataset with text and labels
data = {
    'text': [
        "The phone has great battery life.",  # Document 1 (Tech)
        "This battery lasts long.",           # Document 2 (Tech)
        "Exercise and diet improve health.",  # Document 3 (Health)
        "Good diet keeps you healthy.",       # Document 4 (Health)
        "The phone has a long battery."       # Document 5 (Tech)
    ],
    'label': ['Tech', 'Tech', 'Health', 'Health', 'Tech']
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer


# Initialize the Count Vectorizer
count_vectorizer = CountVectorizer()

# Fit the model and transform the data
count_matrix = count_vectorizer.fit_transform(df.text)
```

# Text Classification

```python
df=pd.read_csv('https://github.com/mohitgupta-1O1/Kaggle-SMS-Spam-Collection-Dataset-/raw/master/spam.csv',encoding = "ISO-8859-1")
```

```python
df
```

```python
X=df.v2
y=df.v1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Initialize the Count Vectorizer
count_vectorizer = CountVectorizer(stop_words='english')

# Fit the model and transform the data
X_train_v = count_vectorizer.fit_transform(X_train)


clf = LogisticRegression()
clf.fit(X_train_v, y_train)
```

```python
X_test.shape
```

```python
X_test_v=count_vectorizer.transform(X_test)
```

```python
y_pred=clf.predict(X_test_v)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

#TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer


# Initialize the TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english')

# Transform the text data
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

```

```python
clf2 = LogisticRegression()
clf2.fit(X_train_tfidf, y_train)
```

```python


# Make predictions
y_pred2 = clf2.predict(X_test_tfidf)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred2)
print(f"Accuracy: {accuracy:.4f}")
```
